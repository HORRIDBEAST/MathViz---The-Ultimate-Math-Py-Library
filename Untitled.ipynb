## New Feature Test Suite (v0.2 scope)

This section tests only the newly added features:

- `LinAlgVisualizer.eigenvector_discovery_mode`
- `StatsVisualizer.hypothesis_tester`
- `StatsVisualizer.bayesian_updater`

Skipped intentionally (already validated earlier):

- `LinAlgVisualizer.space_transformer`
- Dynamic determinant area in `space_transformer`


```python
# Test setup helpers
%matplotlib widget

import importlib
import matplotlib.pyplot as plt
import mathviz
import mathviz.concepts as concepts

# Force-refresh submodule first, then package exports
importlib.reload(concepts)
importlib.reload(mathviz)

from mathviz import LinAlgVisualizer, StatsVisualizer

results = []
LIVE_REFS = []  # Keep viz/fig references alive so widget callbacks stay interactive

def check(name, fn):
    try:
        obj = fn()
        if obj is not None:
            LIVE_REFS.append(obj)
        print(f"✅ {name}")
        results.append((name, True, ""))
    except Exception as e:
        print(f"❌ {name}: {type(e).__name__}: {e}")
        results.append((name, False, f"{type(e).__name__}: {e}"))
```

```python
# 1) LinAlgVisualizer.eigenvector_discovery_mode tests (new feature)

def test_linalg_eigen_basic():
    viz = LinAlgVisualizer()
    fig = viz.eigenvector_discovery_mode(initial_matrix=[[2, 1], [1, 2]])
    assert fig is not None
    assert 'a' in viz.sliders and 'Time (t)' in viz.sliders
    viz.sliders['Time (t)'].set_val(0.25)
    assert 'Eigenvector' in fig.axes[0].get_title()
    return viz, fig

def test_linalg_eigen_invalid_shape_raises():
    viz = LinAlgVisualizer()
    try:
        viz.eigenvector_discovery_mode(initial_matrix=[1, 2, 3])
        raise AssertionError('Expected ValueError for invalid matrix shape')
    except ValueError:
        pass

def test_linalg_eigen_rotation_complex_case():
    # Rotation matrix -> complex eigenvalues over reals
    viz = LinAlgVisualizer()
    fig = viz.eigenvector_discovery_mode(initial_matrix=[[0, -1], [1, 0]])
    title = fig.axes[0].get_title()
    assert 'Eigenvector' in title
    return viz, fig

check('LinAlg eigen mode basic interactivity', test_linalg_eigen_basic)
check('LinAlg eigen mode invalid shape validation', test_linalg_eigen_invalid_shape_raises)
check('LinAlg eigen mode complex-eigenvalue matrix', test_linalg_eigen_rotation_complex_case)

plt.show()
```

```python
# 2) StatsVisualizer.hypothesis_tester tests (new feature)

def test_hypothesis_tester_basic():
    viz = StatsVisualizer()
    fig = viz.hypothesis_tester()
    assert fig is not None
    assert {'sample size n', 'effect size d', 'alpha'}.issubset(set(viz.sliders.keys()))
    assert 'Hypothesis Tester' in fig.axes[0].get_title()
    return viz, fig

def test_hypothesis_tester_edge_alpha_and_n():
    viz = StatsVisualizer()
    fig = viz.hypothesis_tester()
    viz.sliders['alpha'].set_val(0.01)
    viz.sliders['sample size n'].set_val(500)
    viz.sliders['effect size d'].set_val(3.0)
    assert fig is not None
    return viz, fig

check('Stats hypothesis tester basic', test_hypothesis_tester_basic)
check('Stats hypothesis tester edge sliders', test_hypothesis_tester_edge_alpha_and_n)

plt.show()
```

```python
# 3) StatsVisualizer.bayesian_updater tests (new feature)

def test_bayesian_updater_basic():
    viz = StatsVisualizer()
    fig = viz.bayesian_updater()
    assert fig is not None
    assert {'prior α', 'prior β', 'flips n', 'heads k'}.issubset(set(viz.sliders.keys()))
    assert 'Bayesian Updater' in fig.axes[0].get_title()
    return viz, fig

def test_bayesian_updater_edge_values():
    viz = StatsVisualizer()
    fig = viz.bayesian_updater()
    # push priors and data near limits
    viz.sliders['prior α'].set_val(1.0)
    viz.sliders['prior β'].set_val(20.0)
    viz.sliders['flips n'].set_val(200)
    viz.sliders['heads k'].set_val(200)
    assert fig is not None
    return viz, fig

def test_bayesian_updater_heads_gt_flips_guard():
    viz = StatsVisualizer()
    fig = viz.bayesian_updater()
    viz.sliders['flips n'].set_val(10)
    viz.sliders['heads k'].set_val(200)  # method should clamp internally
    assert fig is not None
    return viz, fig

check('Stats Bayesian updater basic', test_bayesian_updater_basic)
check('Stats Bayesian updater edge slider values', test_bayesian_updater_edge_values)
check('Stats Bayesian updater heads>flips guard', test_bayesian_updater_heads_gt_flips_guard)

plt.show()
```

```python
# 4) Summary report

passed = [name for name, ok, _ in results if ok]
failed = [(name, err) for name, ok, err in results if not ok]

print('\n=== New Feature Test Summary ===')
print(f'Total: {len(results)}')
print(f'Passed: {len(passed)}')
print(f'Failed: {len(failed)}')

if failed:
    print('\nFailures:')
    for name, err in failed:
        print(f'- {name}: {err}')
else:
    print('\nAll new-feature tests passed ✅')
```
