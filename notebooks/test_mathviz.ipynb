```python
# ── Cell 4: Mathematical accuracy ─────────────────────────────────────────
import numpy as np
import sympy as sp

x = sp.Symbol("x")

# Derivatives
assert sp.diff(x**3, x)       == 3*x**2,    "❌ power rule"
assert sp.diff(sp.sin(x), x)  == sp.cos(x), "❌ sin derivative"
assert sp.diff(sp.exp(x), x)  == sp.exp(x), "❌ exp derivative"
chain = sp.simplify(sp.diff(sp.sin(x**2), x) - 2*x*sp.cos(x**2))
assert chain == 0,                            "❌ chain rule"
print("✅  Symbolic derivatives: power rule, sin, exp, chain rule")

# Integrals
assert sp.integrate(x**2, (x, 0, 2))       == sp.Rational(8, 3), "❌ ∫x² [0,2]"
assert sp.integrate(x**3, (x, -2, 2))      == 0,                  "❌ odd fn symmetric"
assert sp.integrate(sp.Integer(1),(x,0,5)) == 5,                   "❌ constant integral"
print("✅  Symbolic integrals: power rule, odd function, constant")

# Quadratic vertex / roots
a, b, c = 1.0, -5.0, 6.0   # roots at x=2 and x=3
disc = b**2 - 4*a*c
roots = sorted([(-b + np.sqrt(disc))/(2*a), (-b - np.sqrt(disc))/(2*a)])
assert abs(roots[0] - 2.0) < 1e-9 and abs(roots[1] - 3.0) < 1e-9, "❌ root values"

vertex_x = -b / (2*a)       # vertex of x^2 - 5x + 6 is at x=2.5
assert abs(vertex_x - 2.5) < 1e-9, "❌ vertex x"
print("✅  Quadratic vertex and root formulas")

# Numerical integration accuracy (trapezoidal vs symbolic)
func = sp.lambdify(x, x**2, "numpy")
x_num = np.linspace(0, 1, 10000)
trap = (np.trapezoid if hasattr(np, "trapezoid") else np.trapz)(func(x_num), x_num)
assert abs(trap - 1/3) < 1e-4, "❌ numerical integration off"
print("✅  Trapezoidal numerical integration matches symbolic (∫x² [0,1] = 1/3)")

print("\n✅  All math accuracy checks passed")
```

```python
# ── Cell 5: AlgebraVisualizer ─────────────────────────────────────────────
import matplotlib
matplotlib.use("Agg")
import matplotlib.pyplot as plt
from mathviz import AlgebraVisualizer

viz = AlgebraVisualizer()

# quadratic_explorer — default
fig = viz.quadratic_explorer()
assert fig is not None,          "❌ quadratic_explorer() returned None"
assert "a" in viz.sliders,       "❌ slider 'a' not created"
assert "b" in viz.sliders,       "❌ slider 'b' not created"
assert "c" in viz.sliders,       "❌ slider 'c' not created"
plt.close("all")
print("✅  quadratic_explorer() with default args")

# quadratic_explorer — custom ranges
viz2 = AlgebraVisualizer()
fig2 = viz2.quadratic_explorer(a_range=(-3,3), b_range=(-5,5), c_range=(-4,4),
                                x_range=(-8,8), y_range=(-15,15))
assert fig2 is not None,         "❌ quadratic_explorer with custom ranges returned None"
plt.close("all")
print("✅  quadratic_explorer() with custom ranges")

# polynomial_explorer — all valid degrees
for deg in [2, 3, 4, 5]:
    viz3 = AlgebraVisualizer()
    fig3 = viz3.polynomial_explorer(degree=deg)
    assert fig3 is not None,     f"❌ polynomial_explorer(degree={deg}) returned None"
    plt.close("all")
print("✅  polynomial_explorer() degrees 2–5")

# polynomial_explorer — invalid degrees must raise
for bad in [0, 1, 6]:
    try:
        AlgebraVisualizer().polynomial_explorer(degree=bad)
        print(f"❌  degree={bad} should have raised ValueError")
    except ValueError:
        pass
print("✅  polynomial_explorer() raises ValueError for degrees outside 2–5")
```

```python
# ── Cell 6: CalculusVisualizer ────────────────────────────────────────────
import matplotlib
matplotlib.use("Agg")
import matplotlib.pyplot as plt
from mathviz import CalculusVisualizer

calc = CalculusVisualizer()

# derivative_visualizer
for fn in ["x**2", "x**3 - 3*x**2 + 2*x", "sin(x)", "exp(-x**2)"]:
    f = CalculusVisualizer().derivative_visualizer(fn)
    assert f is not None, f"❌ derivative_visualizer('{fn}') returned None"
    plt.close("all")
print("✅  derivative_visualizer() for poly / trig / exp functions")

# derivative_visualizer — invalid function must NOT crash
f = CalculusVisualizer().derivative_visualizer("totally_invalid_xyz_123")
assert f is not None, "❌ derivative_visualizer should return figure even on bad input"
plt.close("all")
print("✅  derivative_visualizer() handles invalid input gracefully")

# integral_visualizer
for fn, rng in [("x**2", (-1,1)), ("sin(x)", (0, 3.14)), ("x**3", (-2,2))]:
    f = CalculusVisualizer().integral_visualizer(fn, integration_range=rng)
    assert f is not None, f"❌ integral_visualizer('{fn}') returned None"
    plt.close("all")
print("✅  integral_visualizer() for multiple functions and ranges")

# slider must exist after derivative_visualizer
c2 = CalculusVisualizer()
c2.derivative_visualizer("x**2")
assert "x" in c2.sliders, "❌ x-slider not created in derivative_visualizer"
plt.close("all")
print("✅  derivative_visualizer() creates x-slider")
```

```python
# ── Cell 7: MathViz core tools ────────────────────────────────────────────
import matplotlib
matplotlib.use("Agg")
import matplotlib.pyplot as plt
from mathviz import MathViz

# function_explorer
for fn in ["x**2", "sin(x) + cos(2*x)", "exp(-x**2)", "x**4 - 2*x**2 + 1"]:
    f = MathViz().function_explorer(fn, x_range=(-5, 5))
    assert f is not None, f"❌ function_explorer('{fn}') returned None"
    plt.close("all")
print("✅  function_explorer() for various functions")

# multi_function_plotter
f = MathViz().multi_function_plotter(["x**2", "sin(x)", "cos(x)"])
assert f is not None, "❌ multi_function_plotter returned None"
plt.close("all")
try:
    MathViz().multi_function_plotter([])  # must raise
    print("❌  empty list should raise ValueError")
except ValueError:
    print("✅  multi_function_plotter([]) raises ValueError")
plt.close("all")

# parametric_plotter
for xf, yf in [("cos(t)", "sin(t)"), ("sin(3*t)", "sin(2*t)"), ("t", "t**2")]:
    f = MathViz().parametric_plotter(xf, yf, (0, 6.28))
    assert f is not None, f"❌ parametric_plotter('{xf}','{yf}') returned None"
    plt.close("all")
print("✅  parametric_plotter() for circle, Lissajous, and parabolic path")

# save_figure must raise without a figure
try:
    MathViz().save_figure("should_fail.png")
    print("❌  save_figure without figure should raise ValueError")
except ValueError:
    print("✅  save_figure() raises ValueError when no figure exists")
```

```python
# ── Cell 8: export_to_data_url utility ────────────────────────────────────
import matplotlib
matplotlib.use("Agg")
import matplotlib.pyplot as plt
from mathviz.utils import export_to_data_url

fig, ax = plt.subplots()
ax.plot([1, 2, 3], [1, 4, 9])
url = export_to_data_url(fig)

assert isinstance(url, str),                   "❌ return type must be str"
assert url.startswith("data:image/png;base64,"),"❌ wrong data URL prefix"
payload = url.split(",", 1)[1]
assert len(payload) > 200,                     "❌ base64 payload too short"
plt.close("all")
print(f"✅  export_to_data_url() — PNG, {len(payload)} base64 chars")

# JPEG variant
fig2, ax2 = plt.subplots(); ax2.plot([0,1],[0,1])
url_jpg = export_to_data_url(fig2, format="jpg")
assert url_jpg.startswith("data:image/jpg;base64,"), "❌ JPEG prefix wrong"
plt.close("all")
print("✅  export_to_data_url() — JPEG format")

# Two different figures produce different data
fig3, ax3 = plt.subplots(); ax3.plot([1,2],[1,2])
fig4, ax4 = plt.subplots(); ax4.plot([1,2],[2,1])
assert export_to_data_url(fig3) != export_to_data_url(fig4), "❌ distinct figures same URL"
plt.close("all")
print("✅  export_to_data_url() — distinct figures produce distinct URLs")
```
