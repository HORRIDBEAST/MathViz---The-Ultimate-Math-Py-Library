## 1. Installation and Setup

First, let's make sure MathViz is properly installed and check the version:

````python
import mathviz
import numpy as np
%matplotlib widget
import matplotlib.pyplot as plt
from mathviz import AlgebraVisualizer, CalculusVisualizer, MathViz

# Display library information
mathviz.print_info()
print("\nLibrary loaded successfully!")
````

## 2. Basic Algebra Visualizations

Let's start with some basic algebra visualizations. The `AlgebraVisualizer` class provides interactive tools for exploring algebraic functions.

### 2.1 Interactive Quadratic Explorer

Explore how changing coefficients affects quadratic functions:

````python
from mathviz import AlgebraVisualizer

# Create an algebra visualizer
viz = AlgebraVisualizer()

# Interactive quadratic function explorer
# Use the sliders to change coefficients a, b, and c
# Observe how they affect the vertex and roots
viz.quadratic_explorer(
    a_range=(-3, 3),
    b_range=(-5, 5),
    c_range=(-5, 5),
    x_range=(-8, 8),
    y_range=(-15, 15)
)

plt.show()
````

### 2.2 Polynomial Explorer

Explore higher-degree polynomials and see how coefficients affect their shape:

````python
# Explore a 4th degree polynomial
viz.polynomial_explorer(
    degree=4,
    x_range=(-3, 3),
    y_range=(-10, 10)
)

plt.show()
````

## 4. General Purpose Tools

The base `MathViz` class provides general-purpose plotting tools.

````python
# Create the general-purpose visualizer used throughout Section 4 and 7
from mathviz import MathViz
general_viz = MathViz()
print("general_viz ready:", general_viz)
````

### 4.1 Multi-Function Plotter

Compare multiple functions on the same plot:

````python
# Plot several functions on the same axes for easy comparison
general_viz.multi_function_plotter(
    functions=["x**2", "x**3", "2*x", "sin(x)"],
    x_range=(-3, 3)
)
plt.show()
````

### 4.2 Parametric Curves

Create beautiful parametric plots:

````python
# Create a parametric plot (circle)
general_viz.parametric_plotter(
    x_func="cos(t)",
    y_func="sin(t)",
    t_range=(0, 2*np.pi)
)

plt.show()
````

````python
# Create a more complex parametric plot (Lissajous curve)
general_viz.parametric_plotter(
    x_func="sin(3*t)",
    y_func="sin(2*t)",
    t_range=(0, 2*np.pi)
)

plt.show()
````

### 4.3 Function Explorer

General function explorer with automatic feature detection:

````python
# Explore any function with automatic critical point and root detection
general_viz.function_explorer(
    initial_func="x**4 - 2*x**2 + 1",
    x_range=(-3, 3),
    auto_detect_features=True
)

plt.show()
````

## 7. Educational Examples

Here are some educational demonstrations of mathematical concepts:

### 7.1 Quadratic Formula Visualization

See how the discriminant affects the number and type of roots:

````python
# Demonstrate the quadratic formula
viz = AlgebraVisualizer()

# Focus on how discriminant (b² - 4ac) affects roots
viz.quadratic_explorer(
    a_range=(0.5, 2),    # Keep a positive
    b_range=(-4, 4),
    c_range=(-2, 4),     # Vary c to change discriminant
    x_range=(-5, 5),
    y_range=(-3, 8)
)

plt.show()

print("Watch how the roots change:")
print("- When discriminant > 0: Two real roots")
print("- When discriminant = 0: One repeated root")
print("- When discriminant < 0: No real roots")
````

### 7.2 Exploring Trigonometric Functions

Compare sine, cosine, and their transformations:

````python
# Compare trigonometric functions and identities
trig_functions = [
    "sin(x)",
    "cos(x)",
    "sin(x)**2 + cos(x)**2",  # Should equal 1
    "sin(2*x)",
    "2*sin(x)*cos(x)"         # Should equal sin(2x)
]

general_viz.multi_function_plotter(
    functions=trig_functions,
    x_range=(0, 2*np.pi)
)

plt.show()

print("Notice that:")
print("- sin²(x) + cos²(x) = 1 (always equals 1)")
print("- sin(2x) = 2sin(x)cos(x) (they overlap perfectly)")
````

### 7.3 Understanding Derivatives

Visualize the relationship between a function and its derivative:

````python
# Demonstrate derivative concepts
calc_viz.derivative_visualizer(
    func_str="x**3 - 6*x**2 + 9*x + 1",
    x_range=(-1, 5),
    show_function_input=True
)

plt.show()

print("Key observations:")
print("- Where f'(x) = 0, the function has a local maximum or minimum")
print("- Where f'(x) > 0, the function is increasing")
print("- Where f'(x) < 0, the function is decreasing")
print("- The tangent line slope equals f'(x) at any point")
````

## 9. Tips and Best Practices

1. **Run cells in order** — each section may depend on variables created earlier (e.g. `calc_viz`, `general_viz`)
2. **Interactive Elements** — all sliders respond in real-time; click the figure first to activate it
3. **Function Input (scripts)** — any valid SymPy/numpy expression: `sin(x)`, `exp(x)`, `x**2`, `log(x)`, `sqrt(x)`
4. **Function Input (Jupyter notebooks)** — Jupyter's keyboard shortcuts intercept letter keys (e.g. `x` = cut cell) inside Matplotlib text boxes. Use `JupyterSimpleMathViz` for safe text input in notebooks
5. **Range Adjustment** — use range sliders to zoom in/out on interesting regions
6. **Error Handling** — invalid expressions show a red error message instead of crashing; just correct the input
7. **Saving figures** — call `viz.save_figure("filename.png", dpi=300)` on any visualizer
